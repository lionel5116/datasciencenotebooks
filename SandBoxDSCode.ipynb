```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,5,11)
```

```python
y= x ** 2
```

```python
#Object Oriented way
#create the object - figure = an object
fig = plt.figure()
#assign the axes for the figure - L,B,W,H - indicates the size of the graph
axes = fig.add_axes([0.1,0.1,0.8,0.8]) 
axes.plot(x,y)
#set attributes
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Rick James')
```
