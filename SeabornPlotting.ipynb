```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
 #the tips dataset is builtin the seaborn library
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.displot(tips['total_bill'])
```

```python
#joint plot - shows association between total bills and tips
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
#count plot - counts the # of occourences
sns.countplot(x='sex',data=tips)
```

```python
#boxplot,adding a hue as another plot point to measure the days people smoke the most
#In seaborn, the hue parameter represents which column in the data frame, you want to use for color encoding.
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```
